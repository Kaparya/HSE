#### Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.

```python
# Для каждой этнической группы выведите средний балл за экзамен по чтению

%%time

sum_A = 0
count_A = 0
sum_B = 0
count_B = 0
sum_C = 0
count_C = 0
sum_D = 0
count_D = 0
sum_E = 0
count_E = 0

for index, row in df.iterrows():
  if row['race/ethnicity'] == 'group A':
    sum_A += row['reading score']
    count_A += 1
  elif row['race/ethnicity'] == 'group B':
    sum_B += row['reading score']
    count_B += 1
  elif row['race/ethnicity'] == 'group C':
    sum_C += row['reading score']
    count_C += 1
  elif row['race/ethnicity'] == 'group D':
    sum_D += row['reading score']
    count_D += 1
  elif row['race/ethnicity'] == 'group E':
    sum_E += row['reading score']
    count_E += 1

print("group A: {:.2f}".format(sum_A / count_A))
print("group B: {:.2f}".format(sum_B / count_B))
print("group C: {:.2f}".format(sum_C / count_C))
print("group D: {:.2f}".format(sum_D / count_D))
print("group E: {:.2f}".format(sum_E / count_E))
```

```python
# Для каждого уровня образования выведите минимальный балл за экзамен по письму
%%time

min_assoc = 100
min_bach = 100
min_high = 100
min_mast = 100
min_some = 100
min_some_high = 100

for index, row in df.iterrows():
  if row['education'] == 'associate\'s degree':
    min_assoc = min(min_assoc, row['writing score'])
  elif row['education'] == 'bachelor\'s degree':
    min_bach = min(min_bach, row['writing score'])
  elif row['education'] == 'high school':
    min_high = min(min_high, row['writing score'])
  elif row['education'] == 'master\'s degree':
    min_mast = min(min_mast, row['writing score'])
  elif row['education'] == 'some college':
    min_some = min(min_some, row['writing score'])
  elif row['education'] == 'some high school':
    min_some_high = min(min_some_high, row['writing score'])

print("Min for associate\'s degree: {}".format(min_assoc))
print("Min for bachelor\'s degree: {}".format(min_bach))
print("Min for high school: {}".format(min_high))
print("Min for master\'s degree: {}".format(min_mast))
print("Min for some college: {}".format(min_some))
print("Min for some high school: {}".format(min_some_high))
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def check_value(value: int):
  if value >= passmark:
    return "P"
  return "F"

df['Math_PassStatus'] = df['math score'].map(check_value)
print("Количество студентов, сдавших математику: {}".format(len(df[df['Math_PassStatus'] == 'P'])))
print("Количество студентов, не сдавших математику: {}".format(len(df[df['Math_PassStatus'] == 'F'])))

df['Reading_PassStatus'] = df['reading score'].map(check_value)
print("\nКоличество студентов, сдавших чтение: {}".format(len(df[df['Reading_PassStatus'] == 'P'])))
print("Количество студентов, не сдавших чтение: {}".format(len(df[df['Reading_PassStatus'] == 'F'])))

df['Writing_PassStatus'] = df['writing score'].map(check_value)
print("\nКоличество студентов, сдавших письмо: {}".format(len(df[df['Writing_PassStatus'] == 'P'])))
print("Количество студентов, не сдавших письмо: {}".format(len(df[df['Writing_PassStatus'] == 'F'])))

df.sample(5)
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
  if average_mark > 90:
    return "A"
  if average_mark >= 80:
    return "B"
  if average_mark >= 70:
    return "C"
  if average_mark >= 60:
    return "D"
  if average_mark >= 50:
    return "E"
  return "F"

def CountAverage(row):
  return GetGrade((row['math score'] + row['reading score'] + row['writing score']) / 3)

df['Grade'] = df.apply(CountAverage, axis=1)
df.sample(10)
```
